```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/experiment-1.csv')
print(df)

#no. of rows and columns
rows, columns = df.shape
print(f"\nNumber of rows: {rows}, Number of columns: {columns}")
print("Number of columns:", columns)

#Print column names
print("\nColumn names:", df.columns.tolist())
print("Column Names:")
for col in df.columns:
    print(col)

#Change column name 'Name' to 'FirstName'
df.rename(columns={'Name': 'FirstName'}, inplace=True)
print(f"\nThe updated column names are: {df.columns.tolist()}")

#Print the updated column names vertically'
print("Updated Column Names")
for col in df.columns:
  print(col)

#Print last 5 rows from the bottom
print(f"Last 5 rows from the bottom:\n {df.tail(5)}")

#Print the details of student with the lowest marks
lowest_marks_row = df[df['Marks'] == df['Marks'].min()]
print(f"\nThe details of the student with the lowest marks:\n {lowest_marks_row}")

#print("\nDetails of student with the lowest marks:")............
#print(df.loc[df['Marks'].idxmin()])....................

#Find total marks of all female students
female_students = df[df['Sex'] == 'F']
total_marks_female = female_students['Marks'].sum()
print(f"\nTotal marks of all female students: {total_marks_female}")

#total_female_marks = df[df['Sex'] == 'F']['Marks'].sum()...........
#print(f"\nTotal marks of all female students: {total_female_marks}")..........

#List names of all the male students
male_students = df[df['Sex'] == 'M']
male_student_names = male_students['FirstName']
print("\nNames of all the male students:")
for name in male_student_names:
    print(name)

# male_students = df[df['Sex'] == 'M']['FirstName'].tolist().................
# print("\nNames of all male students:")
# for name in male_students:
#     print(name)........................

#Find mean age of the class
mean_age = df['Age'].mean()
print(f"\nMean Age of the class is: {mean_age}")

#Line plot of marks of the class
df['Marks'].plot(kind='line', x=df['FirstName'], y=df['Marks'], title='Marks of the Class')

plt.plot(df['FirstName'], df['Marks'], marker='o')
plt.title('Marks of the Class')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Find the index of the oldest student in the class
oldest_student_index = df['Age'].idxmax()
oldest_student_name = df.loc[oldest_student_index, 'FirstName']
print(f"\nThe index of the oldest student in the class is: {oldest_student_index} and Name of the student is: {oldest_student_name}")

#Sort and print data based on Marks followed by Grade
sorted_df = df.sort_values(by=['Marks', 'Grade'])
print("\nSorted Data:")
print(sorted_df)

#Change the name 'Nihar' to 'Jason Bourne' in the 'FirstName' column
df.loc[df['FirstName'] == 'Nihar', 'FirstName'] = 'Jason Bourne'
print("\nUpdated DataFrame:")
print(df)

df['FirstName'] = df['FirstName'].replace('Nihar', 'Jason Bourne')
print("\nUpdated DataFrame after renaming 'Nihar' to 'Jason Bourne':")
print(df)

#Change and print the order of columns (Name, Sex, Age, Marks, Grade)
df = df[['FirstName', 'Sex', 'Age', 'Marks', 'Grade']]
print("\nUpdated Column Order:")
print(df)

#Count and print number of students by sex and display with suitable column headers
gender_count = df['Sex'].value_counts().reset_index()
gender_count.columns = ['Gender', 'Couuunt']
print("\nNumber of students by gender:")
print(gender_count)

#Delete and print the row where age=46
df = df[df['Age'] != 46]
print("\nDataFrame after deleting the row where age=46:")
print(df)

df = df.loc[df['Age'] != 46]
print("\nDataFrame after deleting the row where age=46:")
print(df)

#Print the data types of individual columns of the DataFrame
print("\nData Types of Individual Columns:")
print(df.dtypes)

#Convert and print the datatype of the 'Age' column from int to float
df['Age'] = df['Age'].astype(float)
print("\nDataFrame after converting 'Age' column from int to float:")
print(df)

#reate a new column named "UpdatedMarks" with 5.5% more marks than the existing Marks column
df['UpdatedMarks'] = df['Marks'] + (0.055 * df['Marks'])
print("\nDataFrame with 'UpdatedMarks' column:")
print(df)

#Delete the "Marks" column
df = df.drop(columns=['Marks'])
print("\nDataFrame after deleting 'Marks' column:")
print(df)













```
